```r
p <- ggplot(data, aes(attraction)) +
    geom_bar(aes(fill=attraction), width = 0.7) + 
    theme_get() +
    scale_fill_brewer(palette="Dark2") + 
    labs(x = "Attraction", y = "Frequency counts", fill="Attraction")
p
```

```r
addmargins(table(copula, attraction))
assocstats(table(copula, attraction))
lables <- c("FALSE" = "Copula = False", "TRUE" = "Copula = True")
p + facet_wrap(.~copula, labeller = as_labeller(lables)) + 
    labs(title="Case Attraction and Copular Infinitives") #,
        #subtitle="chi² = 16.541 (1) p = 4.7604e-05")
```

```r
b <- ggplot(data, aes(attraction, distance_xy, fill=attraction)) +
    geom_boxplot(notch=T, outlier.shape=15,
             outlier.size=2.4) +
    labs(x="Attraction", y="Distance between X and Y") +
    theme_get() + scale_fill_brewer(palette="Dark2")
b
```

```r
b <- ggplot(data, aes(attraction, distance_xy, fill=attraction)) +
    geom_violin(draw_quantiles = c(0.25, 0.5, 0.75)) +
    labs(x="Attraction", y="Distance between X and Y") +
    theme_get() + scale_fill_brewer(palette="viridis")
b
```

```r
mca <- MCA(data[,c("attraction", "copula", "poss_verb", "personal", "author")], method = "Burt")
```

```r
require("factoextra")
fviz_screeplot(mca, addlabels = TRUE, ylim = c(0, 80), barfill="#1B9E77", barcolor="#1B9E77", main="Eigenvalues") + theme_get()
p <- plot(mca, invisible = "ind", graph.type =  "ggplot",
          col.var=c('#1B9E77', '#1B9E77',
                    '#D95F02', '#D95F02',
                    '#7570B3', '#7570B3',
                    '#E7298A', '#E7298A',
                    '#66A61E', '#66A61E', '#66A61E')) + theme_get()
p
```

```r
addmargins(table(author, attraction))
assocstats(table(author, attraction))
p + facet_wrap(.~author)
```

```r
addmargins(table(poss_verb, attraction))
assocstats(table(poss_verb, attraction))
lables <- c("FALSE" = "Possibility Construction = False", "TRUE" = "Possibility Construction = True")
p + facet_wrap(.~poss_verb, labeller = as_labeller(lables))
```
